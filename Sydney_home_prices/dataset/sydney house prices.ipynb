```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df= pd.read_csv("C:\\Users\\shram\\OneDrive\\Desktop\\SydneyHousePrices.csv")
df
```

```python
df.drop(['Id','postalCode',],axis= 'columns',inplace= True)
df.head()
```

```python
df.dropna(inplace= True)
```

```python
df.sort_values(by='sellPrice', ascending=False).head(10)
```

```python
df2 = df[(df['sellPrice'] >= 100000) & (df['sellPrice'] <= 40000000)]
df3= df2[df2['bed']<=20]
df4= df3[df3['bath']<=20]
df4
```

```python
df4.describe()
```

```python
df4['bed_bath_diff'] = abs(df4['bed'] - df4['bath'])
df5= df4[df4['bed_bath_diff'] <= 10]
df5.head()
```

```python
price_threshold = 5000000  # Adjust this threshold as needed
bedroom_threshold = 10 
df6 = df5[(df5['bed'] <= bedroom_threshold) | (df5['sellPrice'] >= price_threshold)]
```

```python
df6
```

```python
df6.sort_values(by='bed')
```

```python
df6.sort_values(by='sellPrice').head(10)
```

```python
df6.drop('bed_bath_diff', axis= 1)
```

```python
df6['Date'] = pd.to_datetime(df6['Date'])
df6= df6[df6['Date'].dt.year > 2015]
df6.sort_values(by='Date').head(10)
```

```python
try:
    pd.to_datetime(df6['Date'], errors= 'raise')
    print("All dates are in the correct format.")
except ValueError:
    print("Some dates are in the wrong format.")
```

```python
df6['propType'].unique()
```

```python
df6['year'] = df6['Date'].dt.year

df6
```

```python
df7=df6[~((df6['bed'] >= 3) & (df6['sellPrice'] < 500000))]
df7=df7[df7['sellPrice'] > 120000]
df7.sort_values(by='sellPrice').head(10)
```

```python
suburb_dummies = pd.get_dummies(df7.suburb)
proptype_dummies = pd.get_dummies(df7.propType)

df8 = pd.concat([df7.drop('suburb', axis=1), suburb_dummies.drop('Padstow', axis=1)], axis=1)
df9 = df8.drop('propType', axis=1)
df10= df9
df11= df10.drop('bed_bath_diff',axis=1)
df12= df11.drop('Date',axis=1)
df12
```

```python
df13=df12.drop('year',axis=1)
df13
```

```python
x= df13.drop('sellPrice',axis=1)
y= df13.sellPrice
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.model_selection import cross_val_score
model_lin= LinearRegression()
model_ran= RandomForestRegressor(min_samples_split=5,min_samples_leaf=4,n_estimators=50)
model_las= Lasso(alpha=1.0)
X_train,X_test,Y_train,Y_test= train_test_split(x,y,test_size=0.2)
```

```python
model_lin.fit(X_train,Y_train)
model_lin.score(X_test,Y_test)
```

```python
model_las.fit(X_train,Y_train)
model_las.score(X_test,Y_test)
```

```python
#from sklearn.model_selection import GridSearchCV, KFold
param_grid_lin = {}
param_grid_las = {
    'alpha': [0.1, 0.5, 1.0]
}

# Define the number of folds for k-fold cross-validation
n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=True, random_state=42)

# Perform GridSearchCV for each model with k-fold cross-validation
grid_lin = GridSearchCV(LinearRegression(), param_grid_lin, cv=kf)
grid_las = GridSearchCV(Lasso(), param_grid_las, cv=kf)

# Fit the models
grid_lin.fit(X_train, Y_train)
grid_las.fit(X_train, Y_train)

# Get the best parameters and scores
best_params_lin = grid_lin.best_params_
best_score_lin = grid_lin.best_score_

best_params_las = grid_las.best_params_
best_score_las = grid_las.best_score_
results = {
    "Model": ["Linear Regression", "Lasso"],
    "Best Parameters": [best_params_lin, best_params_las],
    "Best Score": [best_score_lin, best_score_las]
}

# Create a DataFrame from the results dictionary
results_df = pd.DataFrame(results)
results_df
```

```python
def predict_price(bed, bath, location):
    loc_index = np.where(x.columns == location)[0]
    
    if len(loc_index) > 0:
        x_new = np.zeros(len(x.columns))
        x_new[0] = bed
        x_new[1] = bath
        x_new[loc_index[0]] = 1
        
        return model_lin.predict([x_new])[0]
    else:
        print("Location not found in training data.")
        return None
```
